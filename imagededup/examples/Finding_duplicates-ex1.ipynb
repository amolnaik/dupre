# Hashing

#### Find duplicates using Perceptual hashing along with scores

```python
from imagededup.methods import PHash

phasher = PHash()
duplicates = phasher.find_duplicates(image_dir=image_dir, scores=True)
```

```python
duplicates
```

```python
img_select = {}
for k, v in duplicates.items():
    if (v != []):
        #img_select.append(v)
        img_select[k] = [x[0] for x in v]
img_select
```

### Plotting duplicates for image file

```python
from imagededup.utils import plot_duplicates
plot_duplicates(image_dir=image_dir, duplicate_map=duplicates, filename=file_name)
```

# CNN

#### Find duplicates using CNN along with scores

```python
from imagededup.methods import CNN

cnn_encoder = CNN()
duplicates_cnn = cnn_encoder.find_duplicates(image_dir=image_dir, scores=True)
```

```python
duplicates_cnn
```

```python
img_select = {}
for k, v in duplicates_cnn.items():
    if (v != []):
        #img_select.append(v)
        img_select[k] = [x[0] for x in v]
img_select
```

```python
img_selected = []
for k, v in img_select.items():
    images = []
    if k not in img_selected:
        images.append(k)
        for vi in v:
            images.append(vi)
            img_selected.append(vi)
        print(images)
```

### Plotting duplicates for image file: 'ukbench00120.jpg'

```python
from imagededup.utils import plot_duplicates
plot_duplicates(image_dir=image_dir, duplicate_map=duplicates_cnn, filename=file_name)
```
